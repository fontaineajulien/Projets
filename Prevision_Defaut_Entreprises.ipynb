## Etape 0 : Import des modules Python et de la base de données

Cette étape préliminaire a pour objectif d'importer l'ensemble des modules dont vous aurez besoin au cours du TP (vous pourrez completer cette partie au fur et à mesure de votre avancement dans les travaux), et les données sur lesquelles vous allez travailler.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier

db = pd.read_csv("Base.csv",sep =";",low_memory=False)
```

## Etape 1 : Présentation des données

Cette première étape vous permet de prendre connaisssance du jeu de données mis à votre disposition. Vous devrez dans le cadre de cette première phase des travaux fournir une vue d'ensemble des informations, de leur taux de remplissage, type...

```python
print(db.shape)
db.head()

# Répartition des classes sur les données
freq = pd.crosstab(index=db["Attr65"],columns="count")
print('\nEffectif :')
print(freq)
print('\nFrequence :')
print(freq/freq.sum())
```

On observe un desequilibre au niveau de la repartition des classes. Nous allons donc devoir tenir compte du poids des classes lors de la phase d'elaboration du modele.

## Etape 2 : Traitement et Nettoyage de la base 

Cette étape est indispensable à l'utilisation des données dans le cadre du testing et de la mise en oeuvre d'un modèle. Elle permet de s'assurer que les tests et calibrations de modèle réalisés s'appuient sur des données de qualité acceptable, mais également d'en réduire la complexité.<br>
Il s'agira donc ici de vérifier le formattage des colonnes, de traiter les données manquantes / abberrantes et outliers ou de supprimer les doublons de lignes et de colonnes. C'est dans le cadre de cette phase que doivent si besoin être traitées les questions de conformité et de confidentialité des données (anonymisation du jeu de données...).<br>
Un bonne compréhension des données et une expertise métier certaine sont généralement nécessaires à la réalisation efficace de ces travaux de traitement et de nettoyage de la base.

```python
# On ote les id et la duree avant bankruptcy
base=db.drop(labels='Attr66',axis=1)
base=base.drop(labels='Attr0',axis=1)
print('Base de donnees (shape) :')
print(base.shape)

#partition en apprentissage/test 
Train,Test=train_test_split(base,train_size = 28930,random_state = 1,stratify = base['Attr65'])
Train.index=list(range(len(Train)))
Test.index=list(range(len(Test)))
print('\nBase de donnees TRAIN (shape) :')
print(Train.shape)
print('\nBase de donnees TEST (shape) :')
print(Test.shape)

# On peut vérifier qu'on a conservé les répartitions sur les données d'apprentissage et test
# Répartition des classes sur les données d'apprentissage
freqTrain = pd.crosstab(index=Train["Attr65"],columns="count")
print('Effectif TRAIN :')
print(freqTrain)
print('Frequence TRAIN :')
print(freqTrain/freqTrain.sum())

# Répartition des classes sur les données de test
freqTest = pd.crosstab(index=Test["Attr65"],columns="count")
print('\nEffectif TEST :')
print(freqTest)
print('\nFrequence TEST :')
print(freqTest/freqTest.sum())


# On remplace les valeurs manquqntes  par la mediane de chaque variable dans chaque classe
Train_Y0 = Train[Train['Attr65'] == 0].convert_objects(convert_numeric=True)
med_0 = Train_Y0.median(axis=0)
for i in range(Train_Y0.shape[1]):
    Train_Y0.ix[:,i] = Train_Y0.ix[:,i].fillna(med_0[i])

Train_Y1 = Train[Train['Attr65'] == 1].convert_objects(convert_numeric=True)
med_1 = Train_Y1.median(axis=0)
for i in range(Train_Y1.shape[1]):
    Train_Y1.ix[:,i] = Train_Y1.ix[:,i].fillna(med_1[i])

Train = pd.concat([Train_Y0,Train_Y1])
print('\nBase de donnees TRAIN (shape) :')
print(Train.shape)

Test_Y0 = Test[Test['Attr65'] == 0].convert_objects(convert_numeric=True)
for i in range(Test_Y0.shape[1]):
    Test_Y0.ix[:,i] = Test_Y0.ix[:,i].fillna(med_0[i])

Test_Y1 = Test[Test['Attr65'] == 1].convert_objects(convert_numeric=True)
for i in range(Test_Y1.shape[1]):
    Test_Y1.ix[:,i] = Test_Y1.ix[:,i].fillna(med_1[i])

Test = pd.concat([Test_Y0,Test_Y1])
print('\nBase de donnees TEST (shape) :')
print(Test.shape)

X_Train = Train.drop(labels='Attr65',axis=1)
X_Test  = Test.drop(labels='Attr65',axis=1)
```

On a conserve la repartition des classes dans nos echantillons TRAIN et TEST.

## Etape 3 : Analyse exploratoire 

Vous aurez essentiellement à résumer la distribution de chaque variable (analyse univariée) ainsi que les relations entre les variables (analyse bivariée). Vous devrez ici faire preuve de créativité pour pouvoir capter les différentes informations pertinentes que renferment vos données et les présenter (sous formes de graphes ou de test statistiques). Vous serez peut être amenés à créer de nouvelles variables représentant des classes de vos anciennes variables ou à faire de l'analyse multivariée (réduction de features à travers l'ACP, l'ACM...).<br>
L'utilisation des modules matplotlib, pyplot et seaborn peuvent être pertinente et vous permettre de faire de beaux graphiques.

```python
################ Analyse en composantes principales ################
sc = StandardScaler()
Z = sc.fit_transform(X_Train)
acp = PCA(svd_solver='full')
print('ACP :')
print(acp)

#calculs
coord = acp.fit_transform(Z)

#nombre de composantes calculées
print("\nNombre d'axes principaux")
print(acp.n_components_) # 6

print('\nVariance expliquee')
print(acp.explained_variance_)
#valeur corrigée
eigval = (6-1)/6*acp.explained_variance_
print('\nValeur corrigee :')
print(eigval)
print('\nProportion de variance expliquee')
print(acp.explained_variance_ratio_)
```

```python
#scree plot de la decroissance des valeurs propres
plt.plot(np.arange(1,64+1),eigval)
plt.title("Scree plot")
plt.ylabel("Eigen values")
plt.xlabel("Factor number")
print('\n')
plt.show()

#cumul de variance expliquée
plt.plot(np.arange(1,64+1),np.cumsum(acp.explained_variance_ratio_))
plt.title("Explained variance vs. # of factors")
plt.ylabel("Cumsum explained variance ratio")
plt.xlabel("Factor number")
print('\n')
plt.show()

###### seuils pour test des bâtons brisés
bs = 1/np.arange(64,0,-1)
bs = np.cumsum(bs)
bs = bs[::-1]
#test des bâtons brisés
print('\n')
print(pd.DataFrame({'Val.Propre':eigval,'Seuils':bs}))


#ANALYSE DES VARIABLES
#le champ components_ de l'objet ACP
print('\n')
print(acp.components_)
#racine carrée des valeurs propres
sqrt_eigval = np.sqrt(eigval)
#corrélation des variables avec les axes
corvar = np.zeros((64,64))
for k in range(64):
 corvar[:,k] = acp.components_[k,:] * sqrt_eigval[k]

#afficher la matrice des corrélations variables x facteurs
print('\n')
print(corvar)
#on affiche pour les deux premiers axes
print('\n')
print(pd.DataFrame({'id':X_Train.columns,'COR_1':corvar[:,0],'COR_2':corvar[:,1]}))
```

```python
#cercle des corrélations
fig, axes = plt.subplots(figsize=(8,8))
axes.set_xlim(-1,1)
axes.set_ylim(-1,1)
#affichage des étiquettes (noms des variables)
for j in range(64):
 plt.annotate(X_Train.columns[j],(corvar[j,0],corvar[j,1]))

#ajouter les axes
plt.plot([-1,1],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-1,1],color='silver',linestyle='-',linewidth=1)

#ajouter un cercle
cercle = plt.Circle((0,0),1,color='blue',fill=False)
axes.add_artist(cercle)
#affichage
print('\n')
plt.show()
```

```python
cos2var = corvar**2
print('\n Cosinus carre des variables :')
print(pd.DataFrame({'id':X_Train.columns,'COS2_1':cos2var[:,0],'COS2_2':cos2var[:,1]}))
#contributions des variables
ctrvar = cos2var
for k in range(64):
 ctrvar[:,k] = ctrvar[:,k]/eigval[k]
#on n'affiche que pour les deux premiers axes
print('\nPour les deux premiers axes :')
print(pd.DataFrame({'id':X_Train.columns,'CTR_1':ctrvar[:,0],'CTR_2':ctrvar[:,1]}))
```

```python
############################ Regression logistique ############################
model = LogisticRegression(class_weight='balanced'  ,penalty='l1', C=1 )

#train a new classifier with selected terms
model.fit(X_Train,Train['Attr65'])

#prediction on the test set
predTrain_1= model.predict(X_Train)

#confusion matrix
mcTrain_1 = metrics.confusion_matrix(Train['Attr65'],predTrain_1)
print('Matrice de confusion Regression logistique TRAIN :')
print(mcTrain_1)

#accuracy rate
print('\nEvaluation Regression logistique TRAIN :')
print(metrics.accuracy_score(Train['Attr65'],predTrain_1))

#prediction on the test set
predTest_1= model.predict(X_Test)

#confusion matrix
mcTest_1 = metrics.confusion_matrix(Test['Attr65'],predTest_1)
print('\nMatrice de confusion Regression logistique TEST :')
print(mcTest_1)

#accuracy rate
print('\nEvaluation Regression logistique TEST :')
print(metrics.accuracy_score(Test['Attr65'],predTest_1))



############################ Gradient boosting ############################
from sklearn.ensemble import GradientBoostingClassifier
model_gb = GradientBoostingClassifier()

#train a new classifier with selected terms
model_gb.fit(X_Train,Train['Attr65'])

#prediction on the train set
predTrain_2= model_gb.predict(X_Train)
predTrain_2

#confusion matrix
mcTrain_2 = metrics.confusion_matrix(Train['Attr65'],predTrain_2)
print('\nMatrice de confusion Gradient boosting TRAIN :')
print(mcTrain_2)

#accuracy rate
print('\nEvaluation Gradient boosting TRAIN :')
print(metrics.accuracy_score(Train['Attr65'],predTrain_2))

#prediction on the train set
predTest_2= model_gb.predict(X_Test)
predTest_2

#confusion matrix
mcTest_2 = metrics.confusion_matrix(Test['Attr65'],predTest_2)
print('\nMatrice de confusion Gradient boosting TEST :')
print(mcTest_2)

#accuracy rate
print('\nEvaluation Gradient boosting TEST :')
print(metrics.accuracy_score(Test['Attr65'],predTest_2))



############################ Courbe de roc ############################
scores_logistic=model.predict_proba(X_Test)
fpr_logistic , tpr_logistic , thresholds_logistic=metrics.roc_curve(Test['Attr65'],scores_logistic[:,1],pos_label=1)
roc_auc_logistic=metrics.auc(fpr_logistic,tpr_logistic)

scores_gb=model_gb.predict_proba(X_Test)
fpr_gb , tpr_gb , thresholds_rf=metrics.roc_curve(Test['Attr65'],scores_gb[:,1],pos_label=1)
roc_auc_gb=metrics.auc(fpr_gb,tpr_gb)

plt.figure()
lw=2
plt.plot(fpr_logistic,tpr_logistic,color='darkorange',lw=lw,label='Régression logistique (aire sous la courbe= %0.2f)'%roc_auc_logistic)
plt.plot(fpr_gb,tpr_gb,color='green',lw=lw,label='Gradient boosting (aire sous la courbe= %0.2f)'%roc_auc_gb)

plt.plot([0,1],[0,1],color='navy',lw=lw,linestyle='--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('taux de faux positif')
plt.ylabel('taux de vrai négatifs')
plt.legend(loc="lower right")
print('\n\nCOURBE DE ROC :')
plt.show()
```
